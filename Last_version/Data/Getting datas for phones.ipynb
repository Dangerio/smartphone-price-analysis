```python
import pandas as pd
import numpy as np
df = pd.read_csv('PhonesDF.txt')
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import warnings
warnings.filterwarnings('ignore')

years = []

for j in df['phone_title'][len(years):]:
    l1 = j.split()
    if  l1[-2] + l1[-1] == '128GB' or l1[-2] + l1[-1] == '64GB' or l1[-2] + l1[-1] == '32GB' or l1[-2] + l1[-1] == '16GB' or l1[-2] + l1[-1] == '256GB':
        l1.pop(-2)
    if 'GB' in l1[-2]:
        l1.pop(-2)
    if 'GB' in l1[-1]:
        l1.pop(-1)
        
    idd = '+'.join(l1)

    url = 'https://phonedb.net/index.php?m=device&s=list'

    driver = webdriver.Chrome()
    driver.get(url)
    driver.find_elements(By.TAG_NAME, 'img')[1].click().sendkeys()
    tables = pd.read_html(driver.page_source)
    years.append(int(tables[1][2][0].split()[0].split(',')[0]))
```

```python
df['year_of_release'] = years
```
